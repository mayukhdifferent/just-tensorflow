```python
import tensorflow as tf
import cv2

CATEGORIES = ["Cat", "Dog"]
```

```python
def prepare(filename):
    IMG_SIZE = 100
    img_array = cv2.imread(filename, 0)
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)
```

```python
model = tf.keras.models.load_model("0-64-3-CNN.model")
```

```python
import matplotlib.pyplot as plt

prediction1 = prepare("Test_Img/img1.jpg")
prediction1 = model.predict(prediction1)

print(prediction1)

img_array = cv2.imread("Test_Img/img1.jpg", 0)
plt.title(CATEGORIES[int(prediction1[0][0])])
plt.imshow(img_array, cmap="gray")
plt.show()
```

```python
prediction2 = prepare("Test_Img/img2.jpg")
prediction2 = model.predict(prediction2)

img_array = cv2.imread("Test_Img/img2.jpg", 0)
plt.title(CATEGORIES[int(prediction2[0][0])])
plt.imshow(img_array, cmap="gray")
plt.show()
```
